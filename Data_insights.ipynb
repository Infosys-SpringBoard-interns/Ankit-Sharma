```python
df = df.drop(columns=['BMI'])
df['Obesity_Status']=df['Obesity_Status'].map({'Normal weight':0,'Overweight':1,'Obese':2,'Underweight':3})
df['Blood_Pressure']=df['Blood_Pressure'].map({'Hypertension Stage 1':0,'Hypertension Stage 2':1,'Normal':2,'Elevated':3})
df.head()
```

```python
smallest_age = df['Age'].min()
largest_age = df['Age'].max()
```

```python
print(f"Smallest Age: {smallest_age}")
print(f"Largest Age: {largest_age}")
```

```python
age = [15, 25, 35, 45, 55, 65, 75, 85]  
age_labels = ['15-25', '26-35', '36-45', '46-55', '56-65', '66-75', '76-85']

df['Age_Range'] = pd.cut(df['Age'], bins=age, labels=age_labels)

sns.countplot(x='Age_Range', hue='Obesity_Status', data=df)

plt.show()
```

```python
obesity_mapping = {
    0: "Normal weight",
    1: "Overweight",
    2: "Obese",
    3: "Underweight"
}

print("Mapping of Obesity_Status categories:")
for key, value in obesity_mapping.items():
    print(f"{key} : {value}")
```
